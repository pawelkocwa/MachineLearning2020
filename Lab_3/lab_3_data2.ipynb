### Ex. 1
#### Notice that the scale of the values for each variable is vastly different. A house will typically have 2-5 bedrooms but may have anywhere from hundreds to thousands of square feet. To fix this, we need to do something called “feature normalization”. One way to do this is by subtracting from each value in a feature the mean of that feature, and then dividing by the standard deviation (2 lines of code).

```python
# get data
mean_size = data2['Size'].mean()
mean_bedrooms = data2['Bedrooms'].mean()
mean_price = data2['Price'].mean()

std_size = data2['Size'].std()
std_bedrooms = data2['Bedrooms'].std()
std_price = data2['Price'].std()

print(f"Mean:\n\t-Size: {mean_size};\n\t-Bedrooms: {mean_bedrooms};\n\t-Price: {mean_price}", end="\n\n")
print(f"Standard deviation:\n\t-Size: {std_size};\n\t-Bedrooms: {std_bedrooms};\n\t-Price: {std_price}")
```

```python
data2['Size'] = (data2['Size'] - mean_size)/std_size
data2['Bedrooms'] = (data2['Bedrooms'] - mean_bedrooms)/std_bedrooms
data2['Price'] = (data2['Price'] - mean_price)/std_price
```

```python
""" Steps from task 1 """

# adding ones
# data2.insert(0, '1s', 1.0)
data2.head()

# seperate data into independent variables
X = data2.drop(['Price'], axis=1)
y = data2.drop(['Size', '1s', 'Bedrooms'], axis=1)
X_m = np.matrix(X)
y_m = np.matrix(y)
theta_m = np.matrix([0,0,0])

# cost function
def computeCost(X, y, theta):
    f_x = np.dot(X, theta.transpose())
    sum_of_pow = np.sum(np.power(f_x-y, 2), axis=0)
    ret_val = sum_of_pow.sum() / (2*len(X)) 
    return ret_val
    
# cost function test for theta=0
print(f"Computed cost: {computeCost(X_m, y_m, theta_m)}")

# gradient descet method
def gradient_prosty(X, y, theta, alpha, it):
    # it - iteration nb.   
    cost = np.zeros(it)
     
    for i in range(it):       
        temp1 = np.multiply(theta, X)
        temp2 = temp1.sum(axis = 1)
        
        cost[i] = computeCost(X, y, theta)
        theta = np.matrix(np.array(theta[0] -(alpha/len(X))*np.sum((np.multiply((temp2 - y), X) ), axis = 0)))
        
    return theta, cost

# calculation of the optimal parameters
alpha = 0.01
it = 1000
theta_out, cost_out = gradient_prosty(X_m, y_m, theta_m, alpha, it)
print(f'Theta: {theta_out}')

# calculate cost function
print(f"Computed cost: {computeCost(X_m, y_m, theta_out)}")

# prepare the linear regression line
#    - create min max 
c_x_min, c_x_max, c_y_min, c_y_max = 0.8, 1.2, 1.5, 1.2

#    - create linear regression line
x_reg = np.linspace(X.min(axis=0)*c_x_min, X.max(axis=0)*c_x_max, 2)
y_reg = theta_out[0,1]*x_reg+theta_out[0,0]

# plot it
# plot the regression function and dataset
plt.plot(X[["Bedrooms", "Size"]], y, '.')
plt.plot(x_reg, y_reg, 'r')
plt.grid()
plt.title("Data with the regression line")
plt.xlabel("Size, Bedrooms")
plt.ylabel('Price')
plt.legend(["P(Bedrooms)", "P(Size)", "Regression line"])

plt.figure()
plt.plot(np.linspace(0, len(cost_out), len(cost_out)), cost_out, 'b--')
plt.grid(True)
plt.title("Relationship between the cost function and number of iterations")
plt.xlabel("Iterations")
plt.ylabel('Cost function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model as linm
 
# Regression models
# http://scikit-learn.org/stable/modules/linear_model.html
 
# Load the diabetes dataset
boston = datasets.load_boston()
# print description
print(boston.DESCR)
# get the data
boston_X = boston.data
boston_Y = boston.target
```

```python
# Import danych
# Normalizacja/Standaryzacja
# Podział na zbiór treningowy i testowy (70-30%)
 
# Stworzenie obiektu 
regr = linear_model.LinearRegression()
 
# Uczenie modelu przy pomocy bazy treningowej
regr.fit(boston_X, boston_Y)
# Przewidywanie wartości dla danych testowych
Y_predicted = regr.predict(boston_X)
 
# Wyświetlenie parametrów prostej
print('Coefficients: \n', regr.coef_)
 
#  Obliczamy rzeczywisty popełniony błąd średnio-kwadratowy
error = np.mean((regr.predict(boston_X) - boston_Y) ** 2)
print("Residual sum of squares: {}".format(error))
```

```python
for i in range(13):
    plt.figure(figsize=(10,10))
    
    plt.plot(boston_X[:,i], boston_Y,'o')
    arg_max = boston_X[:,i].argmax()
    arg_min = boston_X[:,i].argmin()
    
    f_max = np.dot(boston_X[arg_max], regr.coef_) + regr.intercept_
    f_min = np.dot(boston_X[arg_min], regr.coef_) + regr.intercept_
    
    plt.plot([boston_X[arg_min, i], boston_X[arg_max, i]], [f_min, f_max])
    plt.title("X"+str(i))
```
